```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import math
import datetime
```

```python
path = r'C:\Users\Mike\OneDrive\Documents\Rutgers\Senior\Spring\Intro to Data Science\F500_2019'
filenames = glob.glob(os.path.join(path, '*.csv'))

data = []
for csv in filenames:
    frame = pd.read_csv(csv)
    frame['filename'] = os.path.basename(csv)
    data.append(frame)
```

```python
bigframe = pd.concat(data, ignore_index=True)
bigframe['Datetime'] = pd.to_datetime(bigframe['Date'])
bigframe
```

```python
aframe = bigframe
aframe
```

```python
aframe.set_index('Datetime')
```

```python
aframe["HL_PCT"] = (aframe['Adj_High'] - aframe['Adj_Low'])/aframe['Adj_Close'] * 100
aframe["PCTD"] = (aframe['Adj_Close'] - aframe["Adj_Open"])/aframe['Adj_Open'] * 100
aframe
```

```python
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
```

```python
aframe.fillna(-123456789, inplace = True)
aframe['tickerID'] = aframe.groupby(aframe['filename']).ngroup()
aframe
```

```python
frame1 = aframe[(aframe['tickerID'] == 0)]
# & (aframe['Datetime'] < '2020-01-01')
m,n = frame1.shape
x = frame1[['HL_PCT', 'PCTD', 'Volume', 'Dividend', 'Split', 'High']]
y = frame1[['Close']]
date = frame1.loc[:, ['Datetime']]
x_test = x.loc[:np.floor(m*0.3)]
x_train = x.loc[np.floor(m*0.3)+1:]
y_test = y.loc[:np.floor(m*0.3)]
y_train = y.loc[np.floor(m*0.3)+1:]
date_test = date.loc[:np.floor(m*0.3)]
date_train = date.loc[np.floor(m*0.3)+1:]
```

```python
lreg = LinearRegression()
lreg.fit(x_train, y_train)
```

```python
lreg.score(x_test, y_test)
```

```python
np.mean((lreg.predict(x_test) - y_test) **2)
```

```python
plt.xticks(rotation = 45)
plt.plot_date(date_test, y_test, fmt='b-', xdate = True, ydate = False, label = 'Real value')
plt.plot_date(date_test, lreg.predict(x_test), fmt='r-', xdate = True, ydate = False, label = 'Predicted value')
plt.legend(loc = 'upper center')
plt.ylabel('Close prices')
plt.title(frame1['filename'].iloc[0])
```

```python
frame2 = aframe[(aframe['tickerID'] == 0)]
frame2['mavg'] = frame2['Adj_Close'].rolling(window = 120).mean()
frame2 = frame2[frame2['Date'] > '2015-01-01']
plt.plot_date(frame2['Datetime'], frame2['Adj_Close'], fmt='b-', xdate = True, ydate = False, label = 'Real value')
plt.plot_date(frame2['Datetime'], frame2['mavg'], fmt='r-', xdate = True, ydate = False, label = 'Real value')
```

```python
frame3 = aframe[(aframe['tickerID'] == 0)].set_index('Datetime')
frame3.fillna(lambda x : x.median(), inplace = True)
frame3 = frame3.sort_values(by = 'Datetime')
forecast_out = int(math.ceil(0.01 * len(frame3)))
forecast_col = 'Close'

frame3['label'] = frame3[forecast_col].shift(-forecast_out)
X1 = np.array(frame3.drop(['label', 'Date', 'filename'], 1))
X1 = preprocessing.scale(X1)
X1_lately = X1[-forecast_out:]
X1 = X1[:-forecast_out]
y1 = np.array(frame3['label'])
y1 = y1[:-forecast_out]
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2)
lreg2 = LinearRegression(n_jobs = -1)
lreg2.fit(X1_train, y1_train)
```

```python
lreg2.score(X1_test, y1_test)
```

```python
forecast_set = lreg2.predict(X1_lately)
frame3['Forecast'] = np.nan
#last_date = frame3.iloc[-1].name
#last_unix = last_date
#next_unix = last_unix + datetime.timedelta(days=1)

#for i in forecast_set:
#    next_date = next_unix
#    next_unix += datetime.timedelta(days=1)
#    frame3.loc[next_date] = [np.nan for _ in range(len(frame3.columns)-1)]+[i]
    
last_date = datetime.datetime(2020, 5, 8)

for i in forecast_set:
    frame3.loc[last_date] = [np.nan for _ in range(len(frame3.columns)-1)]+[i]
    last_date -= datetime.timedelta(days = 1)
```

```python
plt.xticks(rotation = 45)
plt.plot_date(date_test, y_test, fmt='b-', xdate = True, ydate = False, label = 'Real value')
plt.plot_date(date_test, lreg.predict(x_test), fmt='r-', xdate = True, ydate = False, label = 'Predicted value')
frame3['Forecast'].tail(500).plot(color = 'green')
plt.legend(loc = 'upper center')
plt.ylabel('Close prices')
plt.title(frame1['filename'].iloc[0])
plt.axis(['2020-01-01','2020-05-08', 0, 700])
```
